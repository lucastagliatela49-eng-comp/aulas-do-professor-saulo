## 2. Lendo um CSV como DataFrame

A estrutura principal do Pandas é o **DataFrame** (que é o nome elegante para "Tabela").

```python
# Vamos criar um CSV básico aqui primeiro (ou ler o turma.csv da aula passada)
import csv
with open("loja_teste.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerows([
        ["Produto", "Preco", "Quantidade"],
        ["Caderno", 15.50, 40],
        ["Caneta", 2.50, 150],
        ["Mochila", 89.90, 10]
    ])

# Lendo a tabela inteira com 1 linha de Pandas!
df = pd.read_csv("loja_teste.csv")
df  # No Jupyter, é só digitar a variável no final da célula para mostrar a tabela bonitinha
```

## 3. Explorando os Dados

Com a tabela em mãos (`df`), o Pandas tem várias "habilidades" mágicas.

```python
print("1. Quantidade de Linhas e Colunas:", df.shape)
print("\n2. Ver os primeiros itens:")
display(df.head(2))
print("\n3. Descrição Estatística (Média, Max, Min):")
display(df.describe())
```

## 4. Manipulando Colunas e Filtros

Para pegar uma coluna inteira, usamos como se fosse a chave de um Dicionário: `df['Preco']`.

```python
# O Pandas é vetorizado. Ele deixa fazer conta na coluna INTEIRA de uma só vez
df['Faturamento_Bruto'] = df['Preco'] * df['Quantidade']

print("Tabela com a nova coluna Matemática:")
display(df)

# Filtrando!
print("\nMostrando APENAS produtos que custam menos de 30 reais:")
display(df[df['Preco'] < 30])
```

### Desafio 1: Tabela de Notas
Crie na mão um pequeno CSV (usando o módulo nativo ou pelo Bloco de Notas) contendo `Aluno, Portugues, Matematica`. Em seguida:
1. Use `pd.read_csv` para carregá-lo.
2. Crie uma coluna nova `Media` como a soma das duas dividida por 2.
3. Filtre na tela apenas os alunos com média >= 7.

```python

# Importe o pandas (se precisar)

import csv
import pandas as pd

dados_alunos = [
    ['Aluno', 'Portugues', 'Matematica'],
    ['Ana', 8, 9],
    ['Bruno', 6, 7.5],
    ['Carla', 9, 8.5],
    ['Daniel', 5.5, 6],
    ['Eduarda', 7, 7]
]

nome_arquivo_csv = 'alunos.csv'


try:
    with open(nome_arquivo_csv, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerows(dados_alunos)
    print(f"Arquivo '{nome_arquivo_csv}' criado com sucesso!")
except Exception as e:
    print(f"Ocorreu um erro ao criar o arquivo CSV: {e}")

# Carregue o CSV aqui

try:
    df_alunos = pd.read_csv(nome_arquivo_csv)
    print(f"Conteúdo do arquivo '{nome_arquivo_csv}' carregado no DataFrame:\n")
    display(df_alunos)
except FileNotFoundError:
    print(f"Erro: O arquivo '{nome_arquivo_csv}' não foi encontrado.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo CSV: {e}")

# Calcule e Média e filtre

df_alunos['Media'] = (df_alunos['Portugues'] + df_alunos['Matematica']) / 2
print("DataFrame com a nova coluna 'Media':\n")
display(df_alunos)

df_alunos_aprovados = df_alunos[df_alunos['Media'] >= 7]
print("Alunos com média maior ou igual a 7:\n")
display(df_alunos_aprovados)
```
