## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

## Exemplo Resolvido 2: Calculadora Simples

**O problema:** Peça dois números e um operador matemático (`+`, `-`, `*` ou `/`) ao usuário. Imprima o resultado da operação.

```python
num1 = float(input("Digite o primeiro número: "))
op = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

if op == "+":
    print(f"Resultado: {num1 + num2}")
elif op == "-":
    print(f"Resultado: {num1 - num2}")
elif op == "*":
    print(f"Resultado: {num1 * num2}")
elif op == "/":
    # Cuidados extras em programação: não se divide por zero!
    if num2 == 0:
        print("Erro: Divisão por zero.")
    else:
        print(f"Resultado: {num1 / num2}")
else:
    print("Operador inválido!")
```

### Desafio 1: Vogal ou Consoante
Faça um programa que verifique se uma letra digitada é vogal ou consoante. Lembre-se, vogais são `a`, `e`, `i`, `o`, `u`.

### Desafio 2: Aumento de Salário
Faça um programa que calcule o aumento de salário de um funcionário.
- Para salários maiores que R$ 1500,00, aumento de 10%.
- Para salários menores ou iguais a R$ 1500,00, aumento de 15%.
Multiplicar um valor por 1.10 é dar 10% de aumento.

```python
salario = float(input("Qual o salário atual? "))

# Seu código aqui
s15 = salario*1.15
s10 = salario*1.10
if salario <= 1500:
  print(f"Seu salário será de:R${s10:.2f}")
else:
  print(f"Seu salário será de: R${s15:.2f}")
```

### Desafio 3: Triângulos
Peça os 3 lados de um triângulo. O programa deverá informar se os valores formam um triângulo. Caso formem, diga o tipo do triângulo:
- Equilátero: 3 lados iguais;
- Isósceles: 2 lados iguais;
- Escaleno: 3 lados diferentes.
*Dica: Para formar um triângulo, a soma de quaisquer dois lados precisa ser maior que o terceiro lado.*

```python
# Seu código aqui

print("Verificador e Classificador de Triângulos")

lado1 = float(input("Digite o comprimento do primeiro lado: "))
lado2 = float(input("Digite o comprimento do segundo lado: "))
lado3 = float(input("Digite o comprimento do terceiro lado: "))


if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Os valores podem formar um triângulo!")

    if lado1 == lado2 and lado2 == lado3:
        print("É um triângulo Equilátero: 3 lados iguais.")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("É um triângulo Isósceles: 2 lados iguais.")
    else:
        print("É um triângulo Escaleno: 3 lados diferentes.")
else:
    print("Os valores NÃO podem formar um triângulo: a soma de quaisquer dois lados precisa ser maior que o terceiro lado.")
```
