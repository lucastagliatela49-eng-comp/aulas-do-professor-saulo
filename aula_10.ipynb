# Aula 10: Arquivos JSON e CSV

Até agora, toda vez que o nosso programa terminava, todas as nossas variáveis (listas, dicionários) eram perdidas.
Para salvar os dados permanentemente (persistência), precisamos aprender a ler e escrever em **Arquivos**.

## 1. Arquivos de Texto Simples `.txt`

O Python possui a função `open()` para abrir arquivos. Precisamos dizer o nome do arquivo e o **Modo** de abertura:
- `'r'` (read): Apenas para Ler (padrão).
- `'w'` (write): Escrever por cima de tudo.
- `'a'` (append): Escrever no final do arquivo.

```python
# Escrevendo em um arquivo (vai ser criado nesta mesma pasta)
arquivo = open("meu_diario.txt", "w")
arquivo.write("Hoje comecei a aprender sobre arquivos em Python.\n")
arquivo.close() # SEMPRE feche o arquivo quando terminar!
```

### O bloco `with`
Esquecer o `arquivo.close()` é comum e pode corromper os dados. A forma profissional de fazer isso no Python é usando o bloco `with`. Ele abre e **garante que vai fechar**.

```python
# Lendo o arquivo usando with (Muito mais seguro e limpo!)
with open("meu_diario.txt", "r") as arquivo:
    conteudo = arquivo.read()
    print("Li do arquivo o seguinte texto:\n" + conteudo)

# Já está fechado aqui fora!
```

## 2. Arquivos JSON

O `.txt` é muito básico. E se quisermos salvar um Dicionário ou uma Lista complexa?
O formto **JSON** (JavaScript Object Notation) é o padrão de mercado para isso na internet. Ele é basicamente texto, mas formatado exatamente como os Dicionários e Listas do Python!

```python
import json

meus_dados = {
    "nome": "Saulo",
    "habilidades": ["Python", "Dados"],
    "nivel": 10
}

# Salvando um Dicionário Python para um arquivo JSON (DUMP)
with open("perfil.json", "w") as f:
    json.dump(meus_dados, f, indent=4)

print("Arquivo JSON criado! Abra a pasta e dê um duplo clique no perfil.json para olhar.")
```

```python
# Lendo um JSON devolta para o Python (LOAD)
with open("perfil.json", "r") as f:
    dados_carregados = json.load(f)

print(f"Bem-vindo de volta, {dados_carregados['nome']}!")
```

## 3. Arquivos CSV (Excel / Planilhas)

CSV (Comma Separated Values) é o formato padrão para exportar tabelas. São como planilhas do Excel, onde cada linha do arquivo é uma linha da tabela, e as colunas são separadas por vírgula `,` ou ponto-e-vírgula `;`.
O Python possui o módulo nativo `csv` para ajudar.

```python
import csv

# Escrevendo dados tabulares
alunos = [
    ["Nome", "Idade", "Nota"],
    ["Maria", 22, 9.5],
    ["José", 25, 7.0]
]

with open("turma.csv", "w", newline='', encoding='utf-8') as f:
    escritor = csv.writer(f)
    escritor.writerows(alunos)

print("Arquivo da turma criado!")
```

### Desafio 1: Agenda Telefônica JSON
Crie um dicionário vazio `agenda = {}`. Peça ao usuário em um loop `while` o nome e o telefone, e os guarde na agenda até ele digitar "sair". No final, use `json.dump` para salvar a agenda em um arquivo `agenda.json`.

```python
import json

agenda = {}

# Seu código aqui

print("Gerenciador de Contatos")
print("Digite 'sair' a qualquer momento para finalizar e salvar a agenda.")

while True:
    nome = input("Digite o nome do contato (ou 'sair' para encerrar): ").strip()
    if nome.lower() == 'sair':
        break

    telefone = input(f"Digite o telefone de {nome}: ").strip()

    agenda[nome] = telefone
    print(f"Contato '{nome}' adicionado.\n")

print("Finalizando o gerenciador de contatos.")

try:
    with open('agenda.json', 'w', encoding='utf-8') as f:
        json.dump(agenda, f, indent=4, ensure_ascii=False)
    print("Agenda salva com sucesso em 'agenda.json'.")
    print("Conteúdo da agenda:")
    print(json.dumps(agenda, indent=4, ensure_ascii=False))
except Exception as e:
    print(f"Ocorreu um erro ao salvar a agenda: {e}")
```

### Desafio 2: Cadastro de Produtos em CSV
Foi te pedida a criação de um sisteminha básico de controle de estoque.
Crie uma lista de listas chamada `produtos`, por exemplo: `[["Notebook", 3500.0, 5], ["Mouse", 120.0, 15]]`.
Abra um arquivo chamado `estoque.csv` em modo de escrita (`'w'`) e use o módulo `csv.writer` para escrever o cabeçalho `["Produto", "Preco", "Quantidade"]` e logo após, as linhas dos produtos. Ao final, confira se o arquivo foi criado com as vírgulas corretamente.

```python
import csv

# O seu código aqui...

produtos = [
    ["Notebook", 3500.0, 5],
    ["Mouse", 120.0, 15],
    ["Teclado", 250.0, 10],
    ["Monitor", 1200.0, 3]
]

nome_arquivo = 'estoque.csv'

cabecalho = ["Produto", "Preco", "Quantidade"]

try:
    with open(nome_arquivo, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(cabecalho)
        writer.writerows(produtos)
    print(f"Arquivo '{nome_arquivo}' criado com sucesso!")
except Exception as e:
    print(f"Ocorreu um erro ao criar o arquivo: {e}")
```

### Desafio 3: Lendo e Adicionando Dados no CSV
Agora que temos o `estoque.csv` criado:
1. Abra o arquivo em modo de leitura (`'r'`) e use o `csv.reader` para ler e imprimir cada linha na tela com um laço `for`.
2. Depois, abra o **mesmo arquivo** em modo de adição/append (`'a'`) e adicione um novo produto usando `writer.writerow(["Teclado", 80.0, 30])`.
3. (Opcional) Abra o arquivo novamente e imprima para comprovar que o Teclado foi adicionado no final!

```python
import csv

# 1. Leitura do Arquivo
nome_arquivo = 'estoque.csv'

print(f"\n--- Conteúdo inicial de '{nome_arquivo}' ---")
try:
    with open(nome_arquivo, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)
except FileNotFoundError:
    print(f"Erro: O arquivo '{nome_arquivo}' não foi encontrado.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo: {e}")

# 2. Adição (append) de novo produto
import pandas as pd

nome_arquivo = 'estoque.csv'
novo_produto = ["Teclado", 80.0, 30]

try:
    with open(nome_arquivo, 'a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(novo_produto)
    print(f"\nProduto '{novo_produto[0]}' adicionado com sucesso ao '{nome_arquivo}'!")
except Exception as e:
    print(f"Ocorreu um erro ao adicionar o produto: {e}")

# Abrir o arquivo novamente e imprimir para comprovar a adição
print(f"\n--- Conteúdo de '{nome_arquivo}' após a adição ---")
try:
    df_estoque_atualizado = pd.read_csv(nome_arquivo)
    display(df_estoque_atualizado)
except FileNotFoundError:
    print(f"Erro: O arquivo '{nome_arquivo}' não foi encontrado após a adição.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo após a adição: {e}")


```
